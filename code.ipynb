```python
import numpy as np
from skimage import io
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pylab as plt
import warnings

warnings.simplefilter('ignore')
```

A

```python
myimg = io.imread('photo.png')
plt.imshow(myimg)
myimgnew = myimg.reshape(myimg.shape[0]*myimg.shape[1], myimg.shape[2])
```

```python
kmeans5 = KMeans(n_clusters = 5, max_iter = 10000)
kmeans5.fit(myimgnew)
km_labels5 = kmeans5.predict(myimgnew)
km_Ccentres5 = kmeans5.cluster_centers_.astype(int)
km_newimg5 = []
for i in range(len(km_labels5)):
    km_newimg5.append(km_Ccentres5[km_labels5[i]])
km_newimg5 = np.array(km_newimg5).reshape(myimg.shape[0], myimg.shape[1], myimg.shape[2])
plt.imshow(km_newimg5)
```

1

```python
num_components = 50
U, s, V = np.linalg.svd(myimgnew)
reconst_img_50 = np.array(U[:, :num_components]).dot(np.diag(s[:num_components]).dot(np.array(V[:num_components, :])))
reconst_img_50 = reconst_img_50.reshape(myimg.shape[0], myimg.shape[1], myimg.shape[2])
reconst_img_50 = np.round(reconst_img_50)
reconst_img_50 = reconst_img_50.astype(int)
plt.imshow(reconst_img_50)
```

C

```python
irisData = load_iris()
X = irisData.data
y = irisData.target
pcadimensiondown = PCA(n_components = 2)
pcadimensiondown.fit(X, y)
pcatransformeddata = pcadimensiondown.transform(X)
plt.scatter(pcatransformeddata[:, 0], pcatransformeddata[:, 1], c = y)
plt.show()
explainedVar = pcadimensiondown.explained_variance_
explainedVarRatio = pcadimensiondown.explained_variance_ratio_
print("Explained variance (and ratio) for principal component 1: ", explainedVar[0], ",", explainedVarRatio[0])
print("Explained variance (and ratio) for principal component 2: ", explainedVar[1], ",", explainedVarRatio[1])
```

1

```python
logreg = LogisticRegression(max_iter = 100)
logreg.fit(X_train, y_train)
y_test_predict = logreg.predict(X_test)
report_logreg = classification_report(y_test, y_test_predict)
print("Classification report for logistic regression:-")
print(report_logreg)
print("Confusion matrix for logistic regression:-")
print(confusion_matrix(y_test, y_test_predict))
```

2

Linear Kernel

```python
svc_linear = SVC(kernel = 'linear', max_iter = 100)
svc_linear.fit(X_train, y_train)
y_test_predict = svc_linear.predict(X_test)
report_svc_linear = classification_report(y_test, y_test_predict)
print("Classification report for SVC (Linear kernel):-")
print(report_svc_linear)
print("Confusion matrix for SVC (Linear kernel):-")
print(confusion_matrix(y_test, y_test_predict))
```

RBF Kernel

```python
svc_rbf = SVC(kernel = 'rbf', max_iter = 100)
svc_rbf.fit(X_train, y_train)
y_test_predict = svc_rbf.predict(X_test)
report_svcrbf = classification_report(y_test, y_test_predict)
print("Classification report for SVC (RBF kernel):-\n")
print(report_svcrbf)
print("Confusion matrix for SVC (RBF kernel):-")
print(confusion_matrix(y_test, y_test_predict))
```
